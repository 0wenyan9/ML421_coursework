```python
#01 - 奇偶數辨別（Odd or even）

#define font styles
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

#main program
while (True):
    input_var01=input(bcolors.BOLD +"Enter an integer [1-max]: ")    
    try:
        var01=int(input_var01)
    except:
        print(bcolors.WARNING +"Error: Please type a valid number") 
        continue
    if var01<1:
        print(bcolors.WARNING +"Error: Please type a positive number") 
    else:
        odd_yes=var01%2
        odd_text_list=["an even","an odd"]
        output_text=str(var01)+" is "+odd_text_list[odd_yes]+" number."
        print(bcolors.BOLD +output_text)
        break
```

```python
#02 - 填字遊戲（Mad libs game）

#main program
while (True):
    input_var01=input("請問需要建造幾次模板故事 [1-max]: ")    
    try:
        var01=int(input_var01)
    except:
        print(bcolors.WARNING +"Error: Please type a valid number") 
        continue
    if var01<1:
        print(bcolors.WARNING +"Error: Please type a positive number") 
    break

loop_var01=1
while (loop_var01<=var01):
    print("\n")
    print("第"+str(loop_var01)+"次 (共"+input_var01+"次模板故事)")
    input_var02=input("請輸入課程期數: ") 
    input_var03=input("請輸入課程名稱: ") 
    input_var04=input("請輸入姓名: ") 
    input_var05=input("請輸入您的背景(科系或職業): ") 
    input_var06=input("請簡述對於本課程得期望(列出幾點修完課程想得到的技能/對自己未來的規劃): ") 
    print("".join(["-"]*30))
    print("第"+input_var02+"期 - "+input_var03)
    print("姓名： "+input_var04)
    print("學生背景： "+input_var05)
    print("".join(["="]*30))
    print(input_var04+" 修習完課程[ "+input_var03," ]後，")
    print("結業成果獲得: "+input_var06)
    print("Well Done.")
    print("".join(["-"]*30))
    loop_var01=loop_var01+1

```

```python
#03 - 猜數字（Guess the number

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

#main program
import math
import random
var01=math.floor(random.randint(1,100))
guess_time=0
var02=-1
abort_yn=0
print(bcolors.BOLD+"猜一個介於範圍（例如 1 到 100）之間的數字: ")

while ((abort_yn==0)and(var02!=var01)):
    input_var02=input(bcolors.BOLD+"猜多少？ （1 到 100，放棄則輸入 0）: ")    
    try:
        var02=int(input_var02)
    except:
        print(bcolors.WARNING +"Error: Please type a valid number") 
        continue
    if (var02 in range(1,101)):
        if (var02>var01):
            print(bcolors.BOLD+"猜錯了，比"+str(var02)+"小")
        elif (var02<var01):
            print(bcolors.BOLD+"猜錯了，比"+str(var02)+"大")
        guess_time=guess_time+1
    elif (var02 ==0):
        abort_yn=1
    else :
        print(bcolors.WARNING +"Error: Please type a number from 1 to 100") 

if (abort_yn==1):
    print(bcolors.BOLD+"沒猜到!共猜"+str(guess_time)+"次")
if (abort_yn==0):           
    print(bcolors.BOLD+"猜對了!共猜"+str(guess_time)+"次")

```

```python
#06 - 及格名單（Pass list）
math_passed={'柯南','灰原','步美','美環','光彦'}
english_passed={'柯南','灰原','丸尾','野口','步美'}
register= (math_passed | english_passed)
register

#function to print the outputs
def f_print(f_text="default",f_set={}):
    f_text_output=""
    if (len(f_set)>0):
        for loop_var01 in range(0,len(f_set)):
            f_text_output=f_text_output+list(f_set)[loop_var01]+" "
    f_text_output=f_text+":"+f_text_output
    print(f_text_output)
    
#數學及格但英文不及格的同學名單
f_print(f_text="數學及格但英文不及格的同學名單",f_set=math_passed & (register-english_passed))

#數學不及格但英文及格的同學名單
f_print(f_text="數學不及格但英文及格的同學名單",f_set=english_passed & (register-math_passed))

#兩者皆及格名單
f_print(f_text="兩者皆及格名單",f_set=(math_passed & english_passed))

```

```python
#08 - 資料夾管理(shutil)

#define font styles
class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'

import os
import shutil
current_path=os.getcwd()
if os.path.exists(current_path+"/files"):
    shutil.rmtree(current_path+"/files")
os.mkdir(current_path+"/files")

#main program
while (True):
    input_var01=input(bcolors.BOLD +"How many folders would you like to create? [enter an integer 1-10]: ")    
    try:
        var01=int(input_var01)
    except:
        print(bcolors.WARNING +"Error: Please type a valid number") 
        continue
    if not(var01 in range(1,11)):
        print(bcolors.WARNING +"Error: Please type an integer between 1 and 10") 
    else:
        for loop_fi in range (1,var01+1):
            os.mkdir(current_path+"/files/f"+str(loop_fi))
        break

shutil.rmtree(current_path+"/files")
```

```python
#09 - 回文判斷（Is a palindrome）

import numpy

input_var01=input(bcolors.BOLD +"enter text and check whether this is palindrome: ") 
#請使用者輸入一串字串，先轉換字串為統一小寫
text_list=numpy.array(list(input_var01.lower()))
#去除相關標點符號
remove_list = [" ", "-", ".", ",", ":", ";", "!", "?", "'", "\""] 
for loop_var01 in remove_list:
    text_list=text_list[text_list!=loop_var01]
    print(text_list)     

text_reverse=text_list[::-1]
print(text_reverse)
if min(text_list==text_reverse)==True:
    print("{} is a palindrome.".format(input_var01))
else:
    print("{} is not a palindrome.".format(input_var01))
```
